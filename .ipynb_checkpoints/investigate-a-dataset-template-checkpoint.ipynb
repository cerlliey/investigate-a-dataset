### Data Cleaning (Addressing duplicates and Null values!)

```python
# After discussing the structure of the data and any problems that need to be
#   cleaned, perform those cleaning steps in the second part of this section.

# dropping duplicate values
df.drop_duplicates(inplace =True)
```

```python
df.shape
```

```python
#drop columns that are not needed

df.drop(['homepage', 'tagline', 'keywords', 'overview', 'production_companies', 'release_date'], axis=1, inplace=True)

df.head()
```

```python
df['original_title'].duplicated().sum()
```

```python
duplicated_titles = df[df.duplicated(['original_title'])]
duplicated_titles = duplicated_titles.sort_values('original_title')
duplicated_titles
```

```python
df['id'].duplicated().sum()
```

```python
df['imdb_id'].duplicated().sum()
```

```python
df.isnull().sum()
```

```python
duplicated_imdb_id = df[df.duplicated(['imdb_id'])]
duplicated_imdb_id
```

```python
#Dropping all NaN values on columns where they appear
columns_with_null = ['imdb_id','cast','director','genres']
df_cleaning = df.dropna(subset=columns_with_null)
```

```python
df_cleaning.shape
```

```python
df_cleaning.isnull().sum()
```

```python
df_cleaning.duplicated().sum()
```

```python
df_cleaning.head(20)
```

```python
df_cleaning.dtypes
```

```python
#making duplicates to count cast, genres and director columns
max_cast_length  = 0
number_of_cast = []
for index in df_cleaning.index:
    index_length = len(df_cleaning['cast'][index].split("|"))
    number_of_cast.append(index_length)
    if index_length > max_cast_length:
        max_cast_length = index_length
df_cleaning['number_of_cast'] = number_of_cast       
max_cast_length
```

```python
max_director_length  = 0
number_of_directors = []
for index in df_cleaning.index:
    index_length = len(df_cleaning['director'][index].split("|"))
    number_of_directors.append(index_length)
    if index_length > max_director_length:
        max_director_length = index_length
    if max_director_length == 36:
        print(df_cleaning['original_title'][index])
        
df_cleaning['number_of_director'] = number_of_directors
max_director_length
```

```python
df_cleaning.sort_values('number_of_director', ascending = False)
```

```python
df_cleaning.head()
```

```python
max_genre_length  = 0
number_of_genre_category = []
for index in df_cleaning.index:
    index_length = len(df_cleaning['genres'][index].split("|"))
    number_of_genre_category.append(index_length)
    if index_length > max_genre_length:
        max_genre_length = index_length

df_cleaning['genre_count'] = number_of_genre_category
max_genre_length
```

```python
df_cleaning.to_csv('cleaned_movie_data_v1.csv', index = False)
```

### Research Question 1 (What kinds of properties are associated with movies that have high revenues (Focus is on adjusted revenue))

```python
#checking if the number of directors influences the the revenue of the movie

df_1.plot(x="revenue_adj", y="number_of_director", kind="scatter");
```

```python
#checking if the popularity of a movie is associated with the 

df_1.plot(x="revenue_adj", y="popularity", kind="scatter");
```

```python
#checking if the movie length is related to the revenue generated from a particular movie

df_1.plot(x="revenue_adj", y="runtime", kind="scatter");
```

## Findings 

Movies with 5 casts earn the highest revenues. This seems to be the threshold for good revenues for movies

A single director movie has a higher revenue performance chance than movies with higher numbers of directors

The number of genres under which a movie is classified under does not have signiificant impact on its revenue

The popularity of a movie does not have signiificant impact on its revenue

The average vote of a movie does not have signiificant impact on its revenue

The runtime of a movie does not have signiificant impact on its revenue


### Research Question 2  (Which genres are most popular in the last 5 years? )

```python
# Creating a copy of dataset for analysis

import numpy as np

df_2 = df_cleaning.copy()
```

```python
df_2.isnull().sum()
```

```python
#creating duplicates of the dataframe to enable them splitting of the genres column  
df_3 = df_2.query('genre_count == 1')
df_4 = df_2.query('genre_count == 2')
df_5 = df_2.query('genre_count == 3')
df_6 = df_2.query('genre_count == 4')
df_7 = df_2.query('genre_count == 5')
```

```python
df_7.isnull().sum()
```

```python
#splitting the columns 

split_columns = 'genres'

df_3[split_columns] = df_3[split_columns].apply(lambda x: x.split('|')[0])
df_4[split_columns] = df_4[split_columns].apply(lambda x: x.split('|')[1])
df_5[split_columns] = df_5[split_columns].apply(lambda x: x.split('|')[2])
df_6[split_columns] = df_6[split_columns].apply(lambda x: x.split('|')[3])
df_7[split_columns] = df_7[split_columns].apply(lambda x: x.split('|')[4])
```

```python
#df_4 = df_4.dropna(subset=['genres'])
df_7.isnull().sum()
```

```python
#appending all dataframes
df_8 = df_3.append([df_4, df_5, df_6, df_7], ignore_index = True)
```
